```python
def is_valid_wikipedia_link(wikipedia_link):
    try:
        response = requests.get(wikipedia_link, allow_redirects=True)
        if response.status_code // 100 == 2:
            return response.url
        else:
            return False
    except requests.RequestException:
        return False  # Handle exceptions such as network errors

def find_first_substring(words, query):
    for word in words:
        if query in word:
            return word
    return None

def find_first_substring_lowercase(words, query):
    for word in words:
        if word.lower()  in query:
            return word
    return None

def find_first_superstring_lowercase(words, query):
    for word in words:
        if query in word.lower():
            return word
    return None

current_document = []
correct_links = []

for index, row in tqdm(merged_test.iterrows()):
    current_document.append({
                    'id': row['id'],
                    'token': row['token'],
                    'full_mention': row['full_mention'],
                    'wiki_link': row['title_mention'],
                    'wiki_url': row['wiki_url_y']
                })
    
    try:
        if row['token'].startswith('-DOCSTART-'):            
            
            document_df = pd.DataFrame(current_document)
            full_mention_ = document_df[document_df['wiki_url'].notna()]
            full_mention_train = full_mention_[full_mention_['wiki_url'] != '?'].set_index('full_mention')['wiki_url'].to_dict()
            mention_test = full_mention_[full_mention_['wiki_url'] == '?']

            incorrect_links = []

            for mention_index, mention_ in mention_test.iterrows():
                link = mention_.wiki_link
                result = is_valid_wikipedia_link(link)
                if result:

                    correct_links.append((mention_.id,result))
                    full_mention_train[mention_.full_mention] = result
                
                else: 
                    correct_attempt = mention_.full_mention.replace(" ","_")
                    correct_attempt = correct_attempt.replace('U.S.', 'United_States')
                    correct_attempt = correct_attempt.replace('U.N.', 'United_Nations')
                    new_link = f'http://en.wikipedia.org/wiki/{correct_attempt}'
                    result = is_valid_wikipedia_link(new_link)
                    if result:
                        correct_links.append((mention_.id,result))
                        full_mention_train[mention_.full_mention] = result
                    else:
                        lowercase_link = new_link.lower()
                        result = is_valid_wikipedia_link(lowercase_link)
                        if result:
                            correct_links.append((mention_.id,result))
                            full_mention_train[mention_.full_mention] = result
                        else:
                            incorrect_links.append((mention_.id,mention_.full_mention))
                            print(f"The Wikipedia link {new_link} is invalid.")

            if len(incorrect_links)> 0:
                print(full_mention_train.keys())
                for incorrect_id, incorrect_term in incorrect_links:
                    incorrect_term = incorrect_term.split('-')[0]
                    
                    match_word = find_first_substring(full_mention_train.keys(),incorrect_term)
                    if match_word:
                        print(incorrect_term,"was mapped to", match_word)
                        correct_links.append((incorrect_id,full_mention_train[match_word]))
                    
                    else:
                        match_word = find_first_substring_lowercase(full_mention_train.keys(),incorrect_term.lower())
                        if match_word:
                            print(incorrect_term,"was mapped to", match_word)
                            correct_links.append((incorrect_id,full_mention_train[match_word]))
                        else:
                            match_word = find_first_superstring_lowercase(full_mention_train.keys(),incorrect_term.lower())
                            if match_word:
                                print(incorrect_term,"was mapped to", match_word)
                                correct_links.append((incorrect_id,full_mention_train[match_word]))
                            else:
                                print("Can't find match for", incorrect_term)
                        
                print("......")
            current_document = []
    except AttributeError:
        pass
```
